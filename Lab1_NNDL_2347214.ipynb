```python
import numpy as np
import matplotlib.pyplot as plt

# Activation function (step function)
def step_function(x):
    return 1 if x >= 0 else 0

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, inputs):
        sum = np.dot(inputs, self.weights) + self.bias
        return step_function(sum)

    def train(self, training_inputs, labels):
        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights += self.learning_rate * (label - prediction) * inputs
                self.bias += self.learning_rate * (label - prediction)
```

```python
# AND Gate
print("AND Gate:")
and_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
and_labels = np.array([0,0,0,1])

and_perceptron = Perceptron(2)
and_perceptron.train(and_inputs, and_labels)

print("Final weights:", and_perceptron.weights)
print("Final bias:", and_perceptron.bias)

for inputs in and_inputs:
    prediction = and_perceptron.predict(inputs)
    print(f"Input: {inputs}, Prediction: {prediction}")

# Plotting decision boundary for AND gate
plt.figure(figsize=(10, 8))
plt.scatter(and_inputs[:, 0], and_inputs[:, 1], c=and_labels, cmap='coolwarm')
x = np.linspace(-0.5, 1.5, 100)
y = -(and_perceptron.weights[0] * x + and_perceptron.bias) / and_perceptron.weights[1]
plt.plot(x, y, 'g--', label='Decision Boundary')
plt.xlabel('Input 1')
plt.ylabel('Input 2')
plt.title('AND Gate - Perceptron Decision Boundary')
plt.legend()
plt.grid(True)
plt.show()
```

AND Gate Classification:

The perceptron successfully learns the AND gate logic. The weights and bias values change during training to create a decision boundary that correctly separates the input (1,1) from the other inputs.

Questions:

1.How do the weights and bias values change during training for the AND gate?

The weights typically become positive, with the bias becoming negative. This allows the perceptron to output 1 only when both inputs are 1.

2.Can the perceptron successfully learn the AND logic with a linear decision boundary?

Yes, the AND gate is linearly separable, so a single-layer perceptron can successfully learn it.

OR Gate Classification:

The perceptron also successfully learns the OR gate logic. The decision boundary is different from the AND gate, allowing the perceptron to output 1 when at least one input is 1.
Questions:

1.What changes in the perceptron's weights are necessary to represent the OR gate logic?

The weights remain positive, but they're typically smaller than for the AND gate. The bias is usually negative but closer to zero compared to the AND gate.

2.How does the linear decision boundary look for the OR gate classification?

The decision boundary separates the (0,0) input from the other inputs.
